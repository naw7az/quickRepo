### What is Matplotlib?
It is a python plotting library, it allows us to turn our data into figures.

### Why Matplotlib?
- Built on Numpy arrays(python)
- Integrates directly with pandas
- Can create basic or advanced plots
- Simple to use interface.

### Workflow of Matplotlib?
1. Create data
2. Create plot(figure)
3. plot data(axes on figure)
4. customize plot
5. Save/share plot

### There are mainlyn 2 ways to plot data in matplotlib
1. pyplot API (less flexible, method 1 and 2)
2. object-orented API (method 3)



```python
# 1st method
fig = plt.figure() # create a figure
ax = fig.add_subplot() # add some axes
plt.show()
```

```python
# 2nd method
fig = plt.figure() # create a figure
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y) # add some data
plt.show()
```

```python
# 3rd method(recommended by matplotlib)
fig, ax = plt.subplots()
ax.plot(x, [50, 100, 200, 400]); # add some data
type(fig), type(ax)
```

### Matplotlib example workflow

```python
# 1. import matplotlib and get it ready for plotting 
%matplotlib inline
import matplotlib.pyplot as plt

# 2. Prepare data
x = [1, 2, 3, 4]
y = [11, 33, 22, 44]

# 3. setup plot
fig,ax = plt.subplots(figsize=(10, 5)) # figsize = (width, height)

# 4 . plot data
ax.plot(x, y)

# 5. customize plot
ax.set(title='Simple Plot', xlabel='x-axis', ylabel='y-axis')

# 6. Save and show
fig.savefig('./sample-plot.png')
```

## Types of Plot :

1. Line plot
2. Scatter plot
3. Bar plot
4. Histogram plot
5. subplot

```python
# Create some data

x = np.linspace(0, 10, 100) # return 100 random val between 0 to 10
x[:10]
```

```python
# 1. line plot
fig,ax = plt.subplots()
ax.plot(x, x**2); # default is line plot
```

```python
# 2. Scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
fig,ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# 3. Bar graph(from dictionary)
nut_butter = {'Almond butter': 10,
             "Peanut butter": 8,
             "Cashew butter": 12}
fig,ax = plt.subplots()
ax.bar(nut_butter.keys(), nut_butter.values())
ax.set(title="Dan's Nut Butter Store", ylabel="Price ($)");
```

```python
# horizontal bar graph
fig, ax = plt.subplots()
# need LIST as input for horizontal bar
ax.barh(list(nut_butter.keys()), list(nut_butter.values()));
```

```python
# 4. histogram
x = np.random.randn(1000) # return val from normal distribution
fig,ax = plt.subplots()
ax.hist(x);
```

### Two ways to make subplot

```python
# 1st way of Subplot
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

# Plot to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter.keys(), nut_butter.values())
ax4.hist(np.random.randn(1000));
```

```python
# 2nd way of Subplot
fig,ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

# Plot to each diffrent index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter.keys(), nut_butter.values())
ax[1, 1].hist(np.random.randn(1000));
```

### Which one should you use? (pyplot vs OO)

- when plotting quickly(just to check data), then `pyplot` is good.
- but when plotting something advance use `OO method`.

```python
heart_disease.head()
```

```python
# taking data of patients over age 50
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
len(over_50), len(heart_disease)
```

### using pyplot(can't see age in x axis)

```python
# c is for color
over_50.plot.scatter(x='age', y='chol', c='target');
```

### OO method mixed with pyplot(can see age in x axis) 

```python
fig, ax = plt.subplots(figsize=(10, 5))
over_50.plot(kind='scatter', x='age', y='chol', c='target', ax=ax);

# to use range in x axis
# ax.set_xlim([45, 100]);
```

### OO method from Scratch

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'])

# Customize
ax.set(title='Heart Disease and Cholesterol Levels',
      xlabel='Age', ylabel='Cholesterol');

# Add a legend
# grabbing target values from scatter and adding legend to each
ax.legend(*scatter.legend_elements(), title='Target');

# Add a horizontal mean line
ax.axhline(over_50['chol'].mean(), linestyle='--');
```

```python
over_50.head()
```

### Subplot of chol, age thalach

```python
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10),
                              sharex=True) # both axis share x axis

# Add data to axis 0
scatter = ax0.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])
# Customizing axis 0
ax0.set(title='Heart Disease and Cholestrol Levels',
        xlabel='Age', ylabel='Cholestrol')
ax0.legend(*scatter.legend_elements(), title='Target')
ax0.axhline(y=over_50['chol'].mean(), linestyle='--')


# Add data to axis 1
thalach = ax1.scatter(x=over_50['age'],
                     y=over_50['thalach'],
                     c=over_50['target'])
# Customizing axis 1
ax1.set(title='Heart Disease and Max Heart Rate',
       xlabel='Age', ylabel='Max Heart Rate')
ax1.legend(*scatter.legend_elements(), title='Target')
ax1.axhline(y=over_50['thalach'].mean(), linestyle='--')

# Add title to entire figure
fig.suptitle('Heart Disease Analysis', 
             fontsize=16, fontweight='bold');
```

## Customizing Your Plots 2 :

```python
# setting style
plt.style.use('seaborn-whitegrid')

fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'],
                    cmap='plasma') # changes the colour(color maps)

# Customize
ax.set(title='Heart Disease and Cholesterol Levels',
      xlabel='Age', ylabel='Cholesterol');

# Add a legend
# grabbing target values from scatter and adding legend to each
ax.legend(*scatter.legend_elements(), title='Target');

# Add a horizontal mean line
ax.axhline(over_50['chol'].mean(), linestyle='--');
```

```python
# Customizing the y and x limitation

fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10),
                              sharex=True) # both axis share x axis

# Add data to axis 0
scatter = ax0.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'],
                     cmap='plasma')
# Customizing axis 0
ax0.set(title='Heart Disease and Cholestrol Levels',
        xlabel='Age', ylabel='Cholestrol')
ax0.legend(*scatter.legend_elements(), title='Target')
ax0.axhline(y=over_50['chol'].mean(), linestyle='--')

# Change ax0 the x axis limits
ax0.set_xlim([50, 80])



# Add data to axis 1
thalach = ax1.scatter(x=over_50['age'],
                     y=over_50['thalach'],
                     c=over_50['target'],
                     cmap='winter')
# Customizing axis 1
ax1.set(title='Heart Disease and Max Heart Rate',
       xlabel='Age', ylabel='Max Heart Rate')
ax1.legend(*scatter.legend_elements(), title='Target')
ax1.axhline(y=over_50['thalach'].mean(), linestyle='--')

# Change ax1 the x axis limits
ax1.set_xlim([50, 80]) # not necessary since both axis share same x

# Change ax1 the y axis limits
ax1.set_ylim([60, 200]) # not necessary since both axis share same x

# Add title to entire figure
fig.suptitle('Heart Disease Analysis', 
             fontsize=16, fontweight='bold');
```
